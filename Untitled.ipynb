```python
import numpy as np
import pandas as pd

def detect_outlier(data_1):
    outliers=set()
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in range(len(data_1)):
        z_score= (data_1[y] - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.add(y)
    return outliers
```

```python
outliers = set()
for column in df:
    outliers |= detect_outlier(df[column])
    print(len(outliers))
```
